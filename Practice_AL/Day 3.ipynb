## Introduction

````python
# Importing Matplotlib
import matplotlib.pyplot as plt
# Import datetime
from datetime import datetime
````

````python
# Creating a graph
fig, ax = plt.subplots() #figure
fig, bx = plt.subplots() #figure

# Line plot
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
# Scatter plot
bx.scatter([1, 2, 3, 4], [10, 20, 25, 30])
````

````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Adding a title
ax.set_title('Simple plot')
````

````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_title('Simple plot')

# Adding labels
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
````

````python
# Matplotlib with dates
dates = ['01/02/1991', '01/03/1991', '01/04/1991']
temperatures = [32, 35, 34]

# Converting the dates to datetime
converted_dates = []
for date in dates:
    new_date = datetime.strptime(date, '%m/%d/%Y')
    converted_dates.append(new_date)
print(date)

# Creating the plot
fig, ax = plt.subplots()
ax.plot(converted_dates, temperatures)
ax.set_xlabel('Date')
ax.set_ylabel('Temperature')
ax.set_title('Temperatures on Dates')

plt.show()
````

## Pandas Data Visualization

Pandas is a data manipulation library in Python. It provides data structures and functions needed to efficiently manipulate large datasets. Pandas has a built-in data visualization feature that uses Matplotlib to create plots.

##

````python
import pandas as pd

# Creating a DataFrame from data.csv
data = pd.read_csv('data.csv')
````

````python
data.head()
````

````python
# Creating a plot
data.plot(x='Age', y='Score', kind='scatter')

## See Matplotlib
````

Pandas has a built-in `.plot()` method that can be used to create a variety of plots. The `.plot()` method is a wrapper around Matplotlib's `pyplot` module. The `.plot()` method can be used to create line plots, bar plots, histograms, scatter plots, and many other types of plots.

Plot Types:
* `line` : Line plot (default)
* `bar` : Vertical bar plot
* `barh` : Horizontal bar plot
* `hist` : Histogram
* `box` : Boxplot
* `kde` : Kernel Density Estimation plot
* `density` : Same as 'kde'
* `area` : Area plot
* `pie` : Pie plot
* `scatter` : Scatter plot
* `hexbin` : Hexbin plot

````python
# Adding title
data.plot(x='Age', y='Score', kind='scatter', title='Age vs Score')
````

Parameters:
* `x` : The column to use for the x-axis values
* `y` : The column to use for the y-axis values
* `kind` : The type of plot to create
* `figsize` : A tuple (width, height) in inches
* `title` : The title of the plot
* `grid` : Display the grid
* `legend` : Display the legend
* `color` : The color of the plot
* `fontsize` : The font size of the title and labels
* `rot` : The rotation of the x-axis labels
* `xlim` : The x-axis limits
* `ylim` : The y-axis limits
* `logy` : Use a logarithmic scale on the y-axis


````python
# Changing the color
data.plot(x='Age', y='Score', kind='scatter', title='Age vs Score', color='red')
````

### Python Functions

```python
def my_function_name(function_arguments, arg1, arg2):

```

### Python Objects

Python objects are instances of classes. Classes are blueprints for creating objects. Objects have attributes and methods that define their behavior. Attributes are variables that store data, and methods are functions that perform actions.

```python
class MyClass:
    # Runs every time a new object is created
    def __init__(self, attribute1, attribute2):
        """
        Initialize the object with the given attributes
        self is a reference to the object itself
        """
        self.class_attribute1 = attribute1
        self.class_attribute2 = attribute2

    # Function specific to the class
    def my_method(self, arg1, arg2):
        """
        Perform an action using the object's attributes
        """
        return self.class_attribute1 + self.class_attribute2 + arg1 + arg2
```
